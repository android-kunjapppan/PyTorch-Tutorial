# For a large datasets, it's better to divide into samples called Batches.

# Then there will be two for loops for training i.e., one for training and another to loop through the Bacthes


* epoch = =1 --> One forward and Backward pass of all Training Samples

* batch_size = number of training samples in one forward and backward pass

* Number of Iterations = number of passes, each pass using [batch_size] number of samples
      eg., 100 samples, batch_size = 20 -->100/20 = 5 iterations per one 1 epoch 

```
import torch 
import torchvision
from torch.utils.data import Dataset , DataLoader
import numpy as np
import pandas as pd
import math
```

```
class Dataset(Dataset):
  def __init__(self):
    # data loading
    xy = np.loadtxt('/content/sample_data/california_housing_train.csv',delimiter = ',',dtype= np.float32,skiprows =1)
    self.x = torch.from_numpy(xy[:,:8])
    self.y = torch.from_numpy(xy[:,[8]])  # n_samples,1
    self.n_samples = xy.shape[0]

  
  def __getitem__(self, index):
    # dataset[0]
    return self.x[index],self.y[index]
  
  def __len__(self):
    # len(dataset)
    return self.n_samples


dataset = Dataset()
first_data = dataset[0]

features, labels = first_data

print(features,labels)
```

# Using DataLoader

```
dataloader = DataLoader(dataset = dataset, batch_size = 4, shuffle=True,num_workers=2)

# num_workers =2 might make loading faster bcoz it's using multiple sub processes now


# Conveting DataLoader into Iterator Object
dataiter = iter(dataloader)
data = dataiter.next()
features, labels = data
print(features, labels)
```
